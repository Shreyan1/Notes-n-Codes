```python
from PIL import Image, ImageOps
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
digits = load_digits()
```

```python
#display the number of rows and cols in the datsets
digits.data.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3)
#train the model - x_train = array of handwritten digit
#y_train = digit
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
img = Image.open('four.png')
```

```python
#grayscale image
img = img.convert('L')
img
```

```python
#invert the colors, black to white and white to black
img_inv = ImageOps.invert(img)
img_inv
```

```python
plt.matshow(img_inv)
```

```python
#convert image into an array - gives 2D array
arr = np.array(img_inv)
arr
```

```python
#convert above 2D into 1D array for prediction
arr1 = arr.flatten()
arr1
```

```python
#let the model predict which image it is
model.predict([arr1])
```
